```python
import os
import jieba
from apyori import apriori
```

```python
punctuation = '0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ⊙∩～~`!@#$%^&*()_-+={}[]|\:;\"\'<>＜＞≪≫,.?/~·•！@#￥%……&*（）\\\
——-—+=｛【】｝』『「」〖〗|、：；“”‘’《》，。？、/*-+.\t\r\n'
trans = str.maketrans(punctuation,' '*len(punctuation))
```

```python
diagnoses = []
for filename in os.listdir('kidney/'):
    diagnosis = []
    with open('kidney/'+filename,'r',encoding = 'UTF-8-sig') as file:
        content = file.readlines()
        flag = 0
        for line in content:
            if flag == 2:
                if '医生' in line or '医师' in line:
                    break
                if '型糖尿病' in line:
                    continue
                diagnosis.append(line)
                    
            if "西医诊断" in line:
                flag += 1
    diagnosis = ' '.join(diagnosis).translate(trans).split()
    tmp = []
    [tmp.append(item) for item in diagnosis if item not in tmp and len(item) > 1 and '极高危' not in item]
    diagnosis = tmp
    if diagnosis != [] and len(diagnosis)<20:
        diagnoses.append(diagnosis)
```

```python
diagnoses
```

```python
results = list(apriori(transactions = diagnoses,min_support = 0.01,min_confidence=0.75,min_lift = 1))
```

```python
for result in results:
    print(result)
```

```python
with open('syndrome_detected.txt','w',encoding = 'UTF-8-sig') as f:
    for rule in results:
        f.write(str(rule)+'\n\n')
```
