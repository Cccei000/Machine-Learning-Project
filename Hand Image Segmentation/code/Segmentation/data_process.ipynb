```python
import os
import numpy as np
from PIL import Image
from scipy import ndimage
from skimage.io import imread
```

```python
import matplotlib.pyplot as plt
def show_img(img):
    fig = plt.figure()
    ax2 = fig.add_subplot(111)
    ax2.imshow(img,cmap='gray')
    plt.show()
```

```python
imglist = []
dirpath = 'img'
dirlist = os.listdir(dirpath) # 1/2/3/4/5/6/7/8
for dirname in dirlist:
    path = dirpath+'/'+dirname # img/1 and so on
    filelist = os.listdir(path)
    for filename in filelist:
        imgpath = path+'/'+filename # img/1/1 (1).bmp and so on
        file = imread(imgpath)
        file = ndimage.zoom(file,(256/480,256/640))
        file = (file-file.min())/(file.max()-file.min())
        imglist.append(file)
img_file = np.array(imglist)
img_file.tofile('img_float64')
```

```python
labellist1 = []
dirpath = 'D:/Jupyter/hands/label/1'
filelist = os.listdir(dirpath)
for filename in filelist:
    path = dirpath+'/'+filename
    file = imread(path)
    file = ndimage.zoom(file,(256/480,256/640))
    file = np.clip(file,0,1)
    labellist1.append(file)
#done
```
